```python
import numpy as np
from stl import mesh
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

```python
d = mesh.Mesh.from_file("/Users/gimli/cvr/data/Pero_drazka_Laserskenr_ME/20201130_1skenr/20201130_1,6mms_25sn_expauto.stl")
e = mesh.Mesh.from_file("/Users/gimli/cvr/data/Pero_drazka_Laserskenr_ME/20201130_1skenr/20201130_3,3mms_25sn_exp1,5ms.stl")
```

```python
d.vectors.shape
```

```python
d.points.shape
```

```python
dpoints = np.unique(np.concatenate(np.array([datum[1] for datum in d.data])), axis=1)
```

```python
dpoints_norm = dpoints - np.mean(dpoints, axis=0)
```

```python
np.var(dpoints_norm, axis=0)
```

```python
epoints = np.unique(np.concatenate(np.array([datum[1] for datum in e.data])), axis=1)
```

```python
epoints_norm = epoints - np.mean(epoints, axis=0)
```

```python
epoints.shape
```

```python
np.var(epoints_norm, axis=0)
```

```python
rd = np.random.randint(0, dpoints.shape[0], 10000)
```

```python
re = np.random.randint(0, epoints.shape[0], 10000)
```

```python
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(dpoints_norm[rd, 0], dpoints_norm[rd, 1], dpoints_norm[rd, 2], alpha=0.5, s=1)
ax.scatter(epoints_norm[re, 0], epoints_norm[re, 1], epoints_norm[re, 2], alpha=0.5, s=1)
plt.show()
```
